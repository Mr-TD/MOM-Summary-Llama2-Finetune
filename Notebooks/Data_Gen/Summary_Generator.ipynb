```python
## import
import pathlib
import textwrap
import google.generativeai as genai
from IPython.display import Markdown
import pandas as pd
import random
```

```python
## init  & coonfigure
import os
from dotenv import load_dotenv
load_dotenv()
GOOGLE_API_KEY = os.getenv('GOOGLE_API_KEY') 
genai.configure(api_key=GOOGLE_API_KEY)

model = genai.GenerativeModel('gemini-pro')
```

```python
## functions
def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
## generate topics & create dataframe
def Summary(transcript):
    prompt = f"""
            You are a skilled writer tasked with transforming meeting minutes into a captivating story that engages and informs readers. Your goal is to craft a narrative that captures the essence of the meeting while keeping readers entertained and interested throughout. Your story should summarize the key points discussed in the meeting in a compelling and engaging manner, making the information easy to understand and remember make a summarize form of the meeting
            as mention below in example any oly incluce key points,major update, priority tasks, current task, completed task nothing else
            Example : Mr. Smith's getting a check-up, and Doctor Hawkins advises him to have one every year. Hawkins'll give some information about their classes and medications to help Mr. Smith quit smoking..
            {transcript}
            """

    response = model.generate_content(prompt)
    trans = response.text
    return trans
```
